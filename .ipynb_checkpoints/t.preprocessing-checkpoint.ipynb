# ترکیب ستونهای (تگ و توصیف) و پاکسازی ان   

```python
def clean_text(val):

    if type(val) != str:
        val = str(val)

    #Removes HTML tags and urls and ...
    val = re.sub('<.*</[a-b]+>', "", val)
    val = re.sub('http.* |http.*\s', "", val)
    val = re.sub('www.* |www.*\s', "", val)
    val = re.sub("(?s)<ref>.+?</ref>", "", val) # remove reference links
    val = re.sub("(?s)<[^>]+>", "", val) # remove html tags
    val = re.sub("&[a-z]+;", "", val) # remove html entities
    val = re.sub("(?s){{.+?}}", "", val) # remove markup tags
    val = re.sub("(?s){.+?}", "", val) # remove markup tags
    val = re.sub("(?s)\[\[([^]]+\|)", "", val) # remove link target strings
    val = re.sub("(?s)\[\[([^]]+\:.+?]])", "", val) # remove media links
    val = re.sub(r'http\S+', 'URL', val)
    val = re.sub('@[\w_]+', 'USER_MENTION', val)
    val = re.sub('\|LBR\|', '', val)
    val = re.sub('\.\.\.+', '...', val)
    val = re.sub('!!+', '!!', val)

    #Normalize tabs and remove newlines
    no_tabs = str(val).replace('\t', ' ').replace('\n', '')
    
    #Remove all characters except A-Z.
    alphas_only = re.sub("[^a-zA-Z,\s]|[\,]", " ", no_tabs)

    #Normalize spaces to 1
    multi_spaces = re.sub(" +", " ", alphas_only)
    
    #Strip trailing and leading spaces
    no_spaces = multi_spaces.strip()
    
    #Normalize all charachters to lowercase
    clean_text = no_spaces.lower()
    
    #If the text has more than one space (removing single word comments) and one character, write it to the file.
    if len(clean_text) == 0 and clean_text.count(' ') == 0:
        return np.nan
    else: return clean_text
```

```python
stop_words = stopwords.words('english')
def tokenizing(val):
    tokenized = word_tokenize(val)
    #Remove stop words
    filtered_sentence = [w for w in tokenized if not w in stop_words]
    return filtered_sentence

df['Tokenized_Tags'] = df['New_Tags'].map(tokenizing)
df['Tokenized_Tags'] = df['Tokenized_Tags'].astype(str).str.replace(r"[\[\]']", '')
df['Tokenized_Tags'] = df['Tokenized_Tags'].replace(',', ' ', regex=True)
df[['Tokenized_Tags','New_Tags']]
```

```python
CORPUS = df['Tokenized_Tags']


from sklearn.feature_extraction.text import CountVectorizer

def bow_extractor(corpus, ngram_range=(1,1)):
    vectorizer = CountVectorizer(min_df=1, ngram_range=ngram_range)
    features = vectorizer.fit_transform(corpus)
    return vectorizer, features

bow_vectorizer, bow_features = bow_extractor(CORPUS)
features = bow_features.todense()
print (features)

feature_names = bow_vectorizer.get_feature_names()
print (feature_names)
```

```python
import pandas as pd
def display_features(features, feature_names):
    df = pd.DataFrame(data=features,
                      columns=feature_names)
    print (df)

display_features(features, feature_names)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
def tfidf_transformer(bow_matrix):
    transformer = TfidfTransformer(norm='l2',
                                   smooth_idf=True,
                                   use_idf=True)
    tfidf_matrix = transformer.fit_transform(bow_matrix)
    return transformer, tfidf_matrix

import numpy as np
#from feature_extractors import tfidf_transformer
feature_names = bow_vectorizer.get_feature_names()
# build tfidf transformer and show train corpus tfidf features
tfidf_trans, tdidf_features = tfidf_transformer(bow_features)
features = np.round(tdidf_features.todense(), 2)
display_features(features, feature_names)
```

```python
import scipy.sparse as sp
from numpy.linalg import norm
feature_names = bow_vectorizer.get_feature_names()

# compute term frequency
tf = bow_features.todense()
tf = np.array(tf, dtype='float64')
# show term frequencies
display_features(tf, feature_names)
```

```python
df = np.diff(sp.csc_matrix(bow_features, copy=True).indptr)
df = 1 + df # to smoothen idf later

display_features([df], feature_names)
```

```python
# compute inverse document frequencies
total_docs = 1 + len(CORPUS)
idf = 1.0 + np.log(float(total_docs) / df)

display_features([np.round(idf, 2)], feature_names)
```

```python
# compute idf diagonal matrix
total_features = bow_features.shape[1]
idf_diag = sp.spdiags(idf, diags=0, m=total_features, n=total_features)
idf = idf_diag.todense()

print (np.round(idf, 2))
```

```python
# compute tfidf feature matrix
tfidf = tf * idf
# show tfidf feature matrix
display_features(np.round(tfidf, 2), feature_names)
```

```python
# compute L2 norms
norms = norm(tfidf, axis=1)
# print norms for each document
print("L2 Normal form",np.round(norms, 2))
```

```python
# compute normalized tfidf
norm_tfidf = tfidf / norms[:, None]
# show final tfidf feature matrix
display_features(np.round(norm_tfidf, 2), feature_names)
```

```python
df['Tokenized_Tags'] = df['Tokenized_Tags'].astype(str)
def review_to_words( raw_review ):
    # Function to convert a raw review to a string of words
    # The input is a single string (a raw movie review), and 
    # the output is a single string (a preprocessed movie review)
    #
    # ۱٫ Remove HTML
    review_text = BeautifulSoup(raw_review).get_text() 
    #
    # ۲٫ Remove non-letters        
    letters_only = re.sub("[^a-zA-Z]", " ", review_text) 
    #
    # ۳٫ Convert to lower case, split into individual words
    words = letters_only.lower().split()                             
    #
    # ۴٫ In Python, searching a set is much faster than searching
    #   a list, so convert the stop words to a set
    stops = set(stopwords.words("english"))                  
    # 
    # ۵٫ Remove stop words
    meaningful_words = [w for w in words if not w in stops]   
    #
    # ۶٫ Join the words back into one string separated by space, 
    # and return the result.
    return( " ".join( meaningful_words ))

clean_review = review_to_words( df["Tokenized_Tags"][16301] )
print (clean_review)
```

```python
#حال این تابع را برای پاکسازی یک باره ی تمام مجموعه آموزشی می نویسیم

# Get the number of reviews based on the dataframe column size
num_reviews = df['Tokenized_Tags'].size

# Initialize an empty list to hold the clean reviews
clean_df_reviews = []

# Loop over each review; create an index i that goes from 0 to the length
# of the movie review list 
for i in range( 0, num_reviews ):
    # Call our function for each one, and add the result to the list of
    # clean reviews
    clean_df_reviews.append( review_to_words( df["Tokenized_Tags"][i] ) )
df
#print(df['Tag Vectors'].iloc[10000])
```

```python
tags_list = " ".join([" ".join(tags) for tags in df['Tokenized_Tags']])    
```

```python
from gensim.summarization import keywords

keywords = keywords(tags_list)
```

```python
keywords = keywords.split('\n')
```

```python
len(keywords)
```

```python
def frequencies(tags):
    key_frequency = ['0'] * len(keywords)
    for tag in tags:
        if tag in keywords:
            key_frequency[keywords.index(tag)] = '1'
    
    compressed = str()
    for i in range(0, len(key_frequency), 128):
        compressed += str(int(''.join(key_frequency[i:i+128]), 2))
    return int(compressed)
```

```python
df['Binary_Tags'] = df['New_Tags'].map(frequencies)
df['Binary_Tags'][1]
```
